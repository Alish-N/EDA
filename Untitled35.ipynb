```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/customer_transactions.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
numerical_features = ['customer_id', 'product_price', 'customer_age']
df[numerical_features].describe()
```

```python
categorical_features = ['product_category', 'purchase_channel' ]
df[categorical_features].describe()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()  #instance of LabelEncoder
for feature in categorical_features:
    df[feature] = le.fit_transform(df[feature])
    print(df[feature])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
print(df[numerical_features])
```

```python
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

#Count outliers for the current feature
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    total_outliers = outliers.shape[0]
#print total number of outliers

    print(f'Total number of outliers in {feature}: {total_outliers}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# customer age distribution by gender
fig = px.histogram(df, x="customer_age", color="customer_gender", nbins=20,
                   title="Distribution of Customer Ages by Gender",
                   labels={"customer_age": "Customer Age"},
                   barmode="overlay")
fig.update_layout(xaxis_title="Customer Age", yaxis_title="Count", legend_title="Gender")
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='customer_gender', y='customer_age', data=df)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python

fig = px.box(df, x="product_category", y="product_price", color="product_category",
             title="Transaction Value Outliers by Product Category",
             labels={"product_category": "Product Category", "product_price": "Transaction Value"})
fig.update_layout(xaxis_title="Product Category", yaxis_title="Transaction Value", showlegend=False)
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='product_category', hue='purchase_channel', data=df)
plt.title('Product Categories by Purchase Channel')
plt.xticks(rotation=45)
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.legend(title='Purchase Channel')
plt.show()
```

```python
#average transaction value by product category and location
avg_transaction = df.groupby(['product_category', 'customer_location'])['product_price'].mean().reset_index()

# Step 2: Find the location with the highest average transaction value for each product category
# `idxmax()` to get the index of the maximum transaction value in each product category group
highest_transaction = avg_transaction.loc[avg_transaction.groupby('product_category')['product_price'].idxmax()]
# `idxmax()` to get the index of the maximum transaction value in each product category group
# Sorting
highest_transaction = highest_transaction.sort_values(by='product_price', ascending=False).reset_index(drop=True)
top_3_highest_transactions = highest_transaction.head(3)

print("\nTop 3 locations with the highest average transaction value:")
print(top_3_highest_transactions)
```

```python
#average transaction graph
fig = px.bar(avg_transaction, x="product_category", y="product_price", color="customer_location",
             title="Average Transaction Value by Product Category and Location",
             labels={"product_category": "Product Category", "product_price": "Avg Transaction Value"})
fig.update_layout(xaxis_title="Product Category", yaxis_title="Average Transaction Value", legend_title="Customer Location")
fig.show()
```
